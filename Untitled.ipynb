```python
from Prompter import LMPrompter

model_base_ = {
    "user_handle":"<human>:",
    "bot_handle":"<bot>:"
}

prompter = LMPrompter(**model_base_)
```

```python
def QA(question, context=None, dont_know=False):
    if (context is None):
        prompt = prompter.prompt("qa_without_context", question)
    else:
        prompt = prompter.prompt("qa_with_context", question, context)
        
    decode_options = {
        "max_new_tokens":256,
        "penalty_alpha":0.9,
        "temperature":0.01,
        "do_sample":True
    }
    
    return prompt, decode_options
```

```python
%%time

question = "What was the capital of the eastern roman empire?"
prompt, decode = QA(question,dont_know=True)
print(prompt)
post_ = {
    "prompt":prompt,
    "generator_options":decode
}

output = requests.post("http://localhost:8200/generate", json=post_)
print(output.text.strip())
```
